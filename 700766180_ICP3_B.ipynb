Notice how the subplot has multiple figures. Now let's add data to each axes.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for plotting
X = np.linspace(0, 10, 100)
Y = X / 2

favorite_foods = {
    'Pizza': 12.99,
    'Sushi': 18.50,
    'Ice Cream': 5.25
}
foods = list(favorite_foods.keys())
prices = list(favorite_foods.values())

random_scatter_x = np.random.rand(10)
random_scatter_y = np.random.rand(10)

normal_data = np.random.normal(loc=0, scale=1, size=1000)

# Create 2x2 subplots with figsize (10, 5)
fig, axs = plt.subplots(2, 2, figsize=(10, 5))

# Top left: Plot X vs X/2
axs[0, 0].plot(X, Y)
axs[0, 0].set_title('X vs X/2')
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('X/2')

# Top right: Scatter plot of 10 random numbers
axs[0, 1].scatter(random_scatter_x, random_scatter_y, color='orange')
axs[0, 1].set_title('Scatter Plot of Random Points')
axs[0, 1].set_xlabel('Random X')
axs[0, 1].set_ylabel('Random Y')

# Bottom left: Bar graph of favorite foods
axs[1, 0].bar(foods, prices, color='green')
axs[1, 0].set_title('Favorite Food Prices')
axs[1, 0].set_xlabel('Food')
axs[1, 0].set_ylabel('Price ($)')

# Bottom right: Histogram of normally distributed data
axs[1, 1].hist(normal_data, bins=30, edgecolor='black')
axs[1, 1].set_title('Histogram of Normal Distribution')
axs[1, 1].set_xlabel('Value')
axs[1, 1].set_ylabel('Frequency')

# Adjust layout so titles/labels don't overlap
plt.tight_layout()

# Show the plot
plt.show()
```

In the process of turning it to numeric data, let's create another column which adds the total amount of sales and another one which shows what date the car was sold.

Hint: To add a column up cumulatively, look up the cumsum() function. And to create a column of dates, look up the date_range() function.

```python
import pandas as pd

# Example raw data with symbols and cents
car_sales = pd.DataFrame({
    'Make': ['Toyota', 'Honda', 'Ford'],
    'Price': ['$20,000.00', '$18,500.50', '$22,999.99']
})

# Step 1: Remove dollar signs and commas
car_sales['Price'] = car_sales['Price'].str.replace('$', '', regex=False)
car_sales['Price'] = car_sales['Price'].str.replace(',', '', regex=False)

# Step 2: Remove the final two digits (after decimal point)
car_sales['Price'] = car_sales['Price'].str.split('.').str[0]

# Step 3: Convert to numeric (int)
car_sales['Price'] = car_sales['Price'].astype(int)

# View cleaned DataFrame
print(car_sales)
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Example cleaned car_sales data
car_sales = pd.DataFrame({
    'Make': ['Toyota', 'Honda', 'Ford', 'BMW', 'Nissan'],
    'Price': [20000, 18000, 22000, 35000, 21000]
})

# Add 'Total Sales' column (cumulative sum of 'Price')
car_sales['Total Sales'] = car_sales['Price'].cumsum()

# Add 'Sale Date' column with successive dates starting from today
today = pd.to_datetime(datetime.today().date())
car_sales['Sale Date'] = pd.date_range(start=today, periods=len(car_sales), freq='D')

# View the DataFrame
print(car_sales)

```

Now we've got a numeric column (`Total Sales`) and a dates column (`Sale Date`), let's visualize them.

```python
import matplotlib.pyplot as plt

# Plot Sale Date vs Total Sales
car_sales.plot(x='Sale Date', y='Total Sales', kind='line', marker='o')

# Add labels and title
plt.title('Total Sales Over Time')
plt.xlabel('Sale Date')
plt.ylabel('Total Sales ($)')
plt.grid(True)

# Show the plot
plt.show()
```

Now let's try comparing two variables versus the target variable.

More specifially we'll see how age and cholesterol combined effect the target in **patients over 50 years old**.

For this next challenge, we're going to be replicating the plot1:


```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create DataFrame with patients over 50 years old
over_50 = heart_disease[heart_disease['age'] > 50]

# Step 2: Plot histograms for numeric columns of the filtered DataFrame
axes = over_50.plot.hist(
    subplots=True,
    layout=(4, 4),      # adjust based on number of numeric columns
    figsize=(12, 15),
    bins=30,
    edgecolor='black',
    alpha=0.7
)

# Step 3: Customize each subplot by adding a mean line
for ax in axes.flatten():
    # Get column name from the subplot title
    col = ax.get_title()

    # Calculate mean value of the column (if numeric)
    if col in over_50.columns and pd.api.types.is_numeric_dtype(over_50[col]):
        mean_val = over_50[col].mean()
        ax.axvline(mean_val, color='red', linestyle='dashed', linewidth=2)
        ax.legend(['Mean', 'Frequency'])
        ax.set_title(col)  # reset title (optional)

plt.tight_layout()
plt.show()
```

Wonderful, you've changed the style of the plots and the figure is looking different but the dots aren't a very good colour.

Let's change the `cmap` parameter of `scatter()` as well as the `color` parameter of `axhline()` to fix it.

Completing this step correctly should result in a figure which looks like plot3

```python
import matplotlib.pyplot as plt

# Filter patients over 50 (if not already done)
over_50 = heart_disease[heart_disease['age'] > 50]

fig, ax = plt.subplots(figsize=(8, 6))

# Scatter plot: age vs cholesterol with winter colormap
scatter = ax.scatter(
    over_50['age'],
    over_50['cholesterol'],
    c=over_50['cholesterol'],
    cmap='winter',
    edgecolor='black',
    alpha=0.7
)

# Add a red dashed horizontal mean line for cholesterol
mean_chol = over_50['cholesterol'].mean()
ax.axhline(mean_chol, color='red', linestyle='--', linewidth=2, label='Mean Cholesterol')

# Customize plot
ax.set_xlabel('Age')
ax.set_ylabel('Cholesterol')
ax.set_title('Cholesterol Levels by Age (Patients over 50)')
ax.legend()

# Add colorbar
fig.colorbar(scatter, ax=ax, label='Cholesterol')

plt.show()
```
